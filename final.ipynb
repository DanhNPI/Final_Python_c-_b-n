```python
from tkinter import *
from tkinter import ttk
from convertLib import *

lst_conversion = ['Decimal', 'Binary', 'Hex']

def display(event):
    disp = lst_selected.get()
    disp1 = lst_selected1.get()
    lbl_notice['text'] = 'Enter your' + ' ' + str(disp).lower() + ' ' + 'number'
    lbl_notice1['text'] = 'Your' + ' ' + str(disp1).lower() + ' ' + 'number'

def convert(event):
    choice = lst_selected.get()
    choice1 = lst_selected1.get()
    num = num1.get()
    print (choice)
    print (choice1)
    print (num)
    if choice == lst_conversion[0] and choice1 == lst_conversion[1]:
        result = dec2bin(int(num))
    if choice == lst_conversion[0] and choice1 == lst_conversion[2]:
        result = dec2hex(int(num))
    if choice == lst_conversion[1] and choice1 == lst_conversion[2]:
        result = bin2hex(num)
    if choice == lst_conversion[1] and choice1 == lst_conversion[0]:
        result = bin2dec(num)
    if choice == lst_conversion[2] and choice1 == lst_conversion[0]:
        result = hex2dec(num)
    if choice == lst_conversion[2] and choice1 == lst_conversion[1]:
        result = hex2bin(num)
    ent2.delete(0, END)
    ent2.insert(0, result)

def reset(event):
    ent1.delete(0, END)
    ent2.delete(0, END)

def swap(event):
    ent1.delete(0, END)
    ent2.delete(0, END)

    temp = ''
    choice_cur = lst_selected.get()
    choice1_cur = lst_selected1.get()
    temp = choice_cur
    choice_cur = choice1_cur
    choice1_cur = temp

    lst_selected.set(choice_cur)
    lst_selected1.set(choice1_cur)

    lbl_notice['text'] = 'Enter your' + ' ' + str(choice_cur).lower() + ' ' + 'number'
    lbl_notice1['text'] = 'Your' + ' ' + str(choice1_cur).lower() + ' ' + 'number'

def check():
    result = ''
    cur_check = status.get()
    cur_selected = lst_selected1.get()
    bin_num = ent2.get()
    if cur_check == True and cur_selected == 'Binary':
        count = 0
        for i in range(len(bin_num)):
            result += bin_num[i]
            if count == 3:
                result += ' '
                count = 0
            else:
                count += 1
        result = result.rstrip(' ')
    else:
        for i in bin_num:
            if i != ' ':
                result += i
    ent2.delete(0, END)
    ent2.insert(0, result)

root = Tk()
root.title('Number conversion')
root.geometry('365x500')
root.resizable(0,0)

# Tạo nhãn Label
lblFrame = LabelFrame(root, text='Number conversion', bg='#ffff99'
                      ,fg='blue', font=('Consolas', 12), labelanchor='n')
lblFrame.pack(fill=BOTH, expand=True, padx=5, pady=5)

lblAction = LabelFrame(root, text='Action', bg='#ffff99'
                      ,fg='blue', font=('Consolas', 12), labelanchor='n')
lblAction.pack(fill=BOTH, expand=True, padx=5, pady=5)

lblFrom = Label(lblFrame, text='From', font=('Consolas', 12), bg='#ffff99')
lblFrom.grid(row=0, column=0, padx=5, pady=5, sticky='w')

lblTo = Label(lblFrame, text='To', font=('Consolas', 12), bg='#ffff99', anchor='w')
lblTo.grid(row=0, column=3, padx=5, pady=5, sticky='w')

lbl_notice = Label(lblFrame, text = 'Enter your decimal number', font=('Consolas', 12), bg='#ffff99')
lbl_notice.place(x = 5, y = 80)

lbl_notice1 = Label(lblFrame, text = 'Your binary number', font=('Consolas', 12), bg='#ffff99')
lbl_notice1.place(x = 5, y = 160)

# Tạo một Entry
num1 = StringVar()

ent1 = Entry(lblFrame, font=('Consolas', 15), justify='right', width=30, textvariable= num1)
ent1.place(x = 5, y = 120)

ent2 = Entry(lblFrame, font=('Consolas', 15), justify='right', width=30)
ent2.place(x = 5, y = 200)

# Tạo một combobox

lst_selected = StringVar()
lst_selected1 = StringVar()
status = BooleanVar()
status.set(False)

comboBox = ttk.Combobox(lblFrame, textvariable=lst_selected, font=('Consolas', 10))
comboBox['values'] = lst_conversion
comboBox.current(0)
comboBox['state'] = 'readonly'
comboBox.grid(row=1, column=0, columnspan=3, padx=5, pady=5, sticky='w')
comboBox.bind('<<ComboboxSelected>>', display)

comboBox1 = ttk.Combobox(lblFrame, textvariable=lst_selected1, font=('Consolas', 10))
comboBox1['values'] = lst_conversion
comboBox1.current(1)
comboBox1['state'] = 'readonly'
comboBox1.grid(row=1, column=3, columnspan=3, padx=5, pady=5, sticky='w')
comboBox1.bind('<<ComboboxSelected>>', display)

# Tạo checkButton

checkBtn = Checkbutton(lblAction, text='Digit grouping', variable=status, font=('Consolas', 12), bg='#ffff99', command=check)
checkBtn.place(x = 5, y = 100)

# Tạo button convert, reset, swap

btn_convert = Button(lblAction, text='Convert', bg='#00e600', font=('Consolas', 12, 'bold'), fg='#f2f2f2', width=8)
btn_convert.place(x=5, y=10,anchor='nw')
btn_convert.bind('<Button-1>', convert)

btn_reset = Button(lblAction, text='Reset', bg='#737373', font=('Consolas', 12, 'bold'), fg='#f2f2f2', width=8)
btn_reset.place(x = 135, y = 10)
btn_reset.bind('<Button-1>', reset)

btn_swap = Button(lblAction, text='Swap', bg='#737373', font=('Consolas', 12, 'bold'), fg='#f2f2f2', width=8)
btn_swap.place(x=260, y=10)
btn_swap.bind('<Button-1>', swap)
root.mainloop()
```
